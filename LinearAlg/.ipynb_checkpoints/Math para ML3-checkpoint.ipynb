## Matrices

Dado el siguiente sistema de ecuaciones:

$$
2a + 3b = 8\newline
10a + 1b =13
$$

Se puede traducir en matrices como:

$$
\begin{pmatrix}
2 & 3\\
10 & 1
\end{pmatrix}
·
\begin{pmatrix}
a\\
b
\end{pmatrix}
=
\begin{pmatrix}
8\\
13
\end{pmatrix}
$$

Existe una relación muy estrecha entre álgebra lineal, sistemas de ecuaciones, vectores y matrices.

## Cómo las matrices trasforman el espacio

Dada esta matriz (sistema de ecuaciones), podemos multiplicarla por los vectores unitarios:

$$
\begin{pmatrix}
2 & 3\\
10 & 1
\end{pmatrix}
·
\begin{pmatrix}
1\\
0
\end{pmatrix}
=
\begin{pmatrix}
2\\
10
\end{pmatrix}
$$

$$
\begin{pmatrix}
2 & 3\\
10 & 1
\end{pmatrix}
·
\begin{pmatrix}
0\\
1
\end{pmatrix}
=
\begin{pmatrix}
3\\
1
\end{pmatrix}
$$
