## Jugando con vectores

Podemos intenr encajar un modelo para representar el histograma anterio

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

def draw(mu, v):
    sigma = math.sqrt(v)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma))

draw(0,1)
draw(0.3,2)
draw(-0.1,4)

plt.show()
```

La pregunta que tenemos es: ¿Cuales son las mejores **mu** y **v** para representar los datos?

Podemos ver un espacio 2d de **mu** y **v** en el que calculamos el error (Distintas medidas del error)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
from scipy.stats import multivariate_normal

x, y = np.mgrid[-1.0:1.0:30j, -1.0:1.0:30j]

# Need an (N, 2) array of (x, y) pairs.
xy = np.column_stack([x.flat, y.flat])

mu = np.array([0.0, 0.0])

sigma = np.array([.5, .5])
covariance = np.diag(sigma**2)

z = -1*multivariate_normal.pdf(xy, mean=mu, cov=covariance)

# Reshape back to a (30, 30) grid.
z = z.reshape(x.shape)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.set_zlabel('Error')
ax.zaxis.label.set_color('red')

#ax.plot_surface(x,y,z)
ax.plot_wireframe(x,y,z)

plt.xlabel('mu')
plt.ylabel('v')

plt.show()
```

Buscaremos cómo movernos en este espacio y qué pasos podemos dar para reducir el error a 0
