## Matrices

Dado el siguiente sistema de ecuaciones:

$$
2a + 3b = 8\newline
10a + 1b =13
$$

Se puede traducir en matrices como:

$$
\begin{pmatrix}
2 & 3\\
10 & 1
\end{pmatrix}
·
\begin{pmatrix}
a\\
b
\end{pmatrix}
=
\begin{pmatrix}
8\\
13
\end{pmatrix}
$$

Existe una relación muy estrecha entre álgebra lineal, sistemas de ecuaciones, vectores y matrices.

## Cómo las matrices trasforman el espacio

Dada esta matriz (sistema de ecuaciones), podemos multiplicarla por los vectores unitarios:

$$
\begin{pmatrix}
2 & 3\\
10 & 1
\end{pmatrix}
·
\begin{bmatrix}
1\\
0
\end{bmatrix}
=
\begin{bmatrix}
2\\
10
\end{bmatrix}
$$

$$
\begin{pmatrix}
2 & 3\\
10 & 1
\end{pmatrix}
·
\begin{bmatrix}
0\\
1
\end{bmatrix}
=
\begin{bmatrix}
3\\
1
\end{bmatrix}
$$

Estos dos vectores generan un esapcio en el que se cumplen todas las propiedades vistas antes.

## Inversas

Es la matriz que multiplicada por sí misma nos da la identidad.

$$
A · A^{-1} = I
$$

Asi podemos resolver ristemas de ecuaciones de forma sencilla (Seria el mismo concepto de pasar términos de un lado de la ecuación a otro)

$$
\begin{pmatrix}
2&3\\
10&1
\end{pmatrix}
*
\begin{bmatrix}
a\\
b
\end{bmatrix}
=
\begin{bmatrix}
8\\
13
\end{bmatrix}
$$

Esto es lo mismo que:

$$
A * r = s 
$$

Tenemos que:

$$
r = A^{-1} * s
$$

## Gaussian elimination

Gaussian elimination, also known as row reduction, is an algorithm in linear algebra for solving a system of linear equations. It is usually understood as a sequence of operations performed on the corresponding matrix of coefficients. This method can also be used to find the rank of a matrix, to calculate the determinant of a matrix, and to calculate the inverse of an invertible square matrix

Operaciones:

- Swapping two rows,
- Multiplying a row by a nonzero number,
- Adding a multiple of one row to another row.

Using these operations, a matrix can always be transformed into an upper triangular matrix, and in fact one that is in row echelon form. Once all of the leading coefficients (the leftmost nonzero entry in each row) are 1, and every column containing a leading coefficient has zeros elsewhere, the matrix is said to be in reduced row echelon form. This final form is unique; in other words, it is independent of the sequence of row operations used.
